```python
# clase Automovil , atributos, marca, placa, precio
# funciones leerdatos(), mostrardatos()
# crear 3 objetos carrro1, carro2, carro3

class Automovil:
    def __init__(self, marca, placa, precio):
        self.marca = marca
        self.placa = placa
        self.precio = precio

    def leerdatos(self):
        self.marca = input("Ingrese la marca: ")
        self.placa = input("Ingrese la placa: ")
        self.precio = int(input("Ingrese el precio: "))
        return self.marca, self.placa, self.precio

    def mostrardatos(self):
        print("Marca: ", self.marca)
        print("Placa: ", self.placa)
        print("Precio: ", self.precio)
        return self.marca, self.placa, self.precio
# fin de la clase automovil
# creando los objetos carro1, carro2, carro3

print("Carro Numero 1")
carro1 = Automovil("", "", 0)
carro1.leerdatos()
carro1.mostrardatos()
print("-*-*-*-*-*-*-*-*-*-*-*-*-*")
print("Carro Numero 2")
carro2 = Automovil("", "", 0)
carro2.leerdatos()
carro2.mostrardatos()
print("-*-*-*-*-*-*-*-*-*-*-*-*-*")
print("Carro Numero 3")
carro3 = Automovil("", "", 0)
carro3.leerdatos()
carro3.mostrardatos()
print("software desarrollado por Sebastian Farfan NL = 16")
```
